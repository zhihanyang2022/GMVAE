```python
import json
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm_notebook

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from math import pi
```

```python
def open_json_as_nparray(json_fname):
    with open(json_fname, 'r') as json_f:
        return np.array(json.load(json_f))
```

```python
smb_int = open_json_as_nparray('smbWithPath-allLevels-chunks-int.json')
```

```python
char2int_smb = { 
    "X": 0,  # smb unpassable
    "S": 1,  # smb breakable
    "-": 2,
    "?": 3,  # smb question
    "Q": 4,  # smb question
    "E": 5,
    "<": 6,
    ">": 7,
    "[": 8,
    "]": 9,
    "o": 10,
    "P": 11,  # for smb path (from Anurag's email)
}
int2char = {v:k for k, v in char2int_smb.items()}
char2name = {
    'X': 'unpassable',
    'S': 'breakable',
    '-': 'background',
    '?': 'SMB question',
    'Q': 'SMB question',
    'E': 'SMB enemy',
    '<': 'Top-left pipe element',
    '>': 'Top-right pipe element',
    '[': 'Bottom-left pipe element',
    ']': 'Bottom-right pipe element',
    'o': 'Coins',
    'P': 'Player path'
}
```

## Density

```python
chunks_int = open_json_as_nparray('./smba_gmvae_4_10000.json')
comp1_chunks = chunks_int[0:2500]
comp2_chunks = chunks_int[2500:5000]
comp3_chunks = chunks_int[5000:7500]
comp4_chunks = chunks_int[7500:10000]
```

```python
def get_densities(chunks):
    unique_indices = np.unique(chunks)
    densities = {i:[] for i in unique_indices}
    for idx, chunk in tqdm_notebook(enumerate(chunks), total=len(chunks), leave=False):
        for i in unique_indices:
            chunk_mod = chunk == i
            densities[i].append(np.count_nonzero(chunk_mod))
    return [np.mean(v) for k, v in densities.items()]
```

```python
comp1_densities = get_densities(comp1_chunks)
comp2_densities = get_densities(comp2_chunks)
comp3_densities = get_densities(comp3_chunks)
comp4_densities = get_densities(comp4_chunks)
```

```python
df = pd.DataFrame([comp1_densities, comp2_densities, comp3_densities, comp4_densities])
df = df.fillna(0)
df /= df.max()
df.columns = [int2char[i] for i in np.unique(chunks_int)]
df.head()
```

```python
categories = list(df)

fig, axs = plt.subplots(nrows=1, ncols=4, figsize=(15, 5),
                           subplot_kw=dict(polar=True))
axs = axs.flatten()

for i in range(4):
    
    ax = axs[i]
    
    values = df.iloc[i].tolist()
    values += values[:1] # repeat the first value to close the circular graph
    angles = [n / float(len(categories)) * 2 * pi for n in range(len(categories))]
    angles += angles[:1]

    ax.set_xticks(angles)
    ax.set_xticklabels(categories)
    ax.set_yticks([])
    ax.set_rlabel_position(30)

    ax.plot(angles, values, linewidth=1, linestyle='solid')
    ax.fill(angles, values, 'skyblue', alpha=0.5)
    
    ax.set_title(f'Component {i+1}', y=1.08)

plt.suptitle('Density of different mario tiles')
plt.tight_layout()
plt.show()
```

```python
densities = {i:[] for i in np.unique(smb_int)}
sideviews = {i:np.zeros((16, len(smb_int))) for i in np.unique(smb_int)}
birdviews = {i:np.zeros((16, len(smb_int))) for i in np.unique(smb_int)}
unique_indexes = np.unique(smb_int)
for idx, chunk in tqdm_notebook(enumerate(smb_int), total=len(smb_int)):
    for i in unique_indexes:
        chunk_mod = chunk == i
        densities[i].append(np.count_nonzero(chunk_mod))
        sideview = chunk_mod.sum(axis=1)
        birdview = chunk_mod.sum(axis=0)
        sideviews[i][:,idx] = sideview
        birdviews[i][:,idx] = birdview
```

```python
densities = {k:float(np.mean(v)) for k, v in densities.items()}
```

```python
plt.bar(densities.keys(), np.array(list(densities.values())) / (16 * 16))
plt.show()
```

```python
sideviews = {k:np.mean(v, axis=1) for k, v in sideviews.items()}
```

```python
import seaborn as sns
```

```python
fig, axs = plt.subplots(3, 4, figsize=(15, 10))
axs = axs.flatten()
for i, ax in enumerate(axs):
    sns.barplot(sideviews[i], np.arange(0, 16), orient='h', ax=ax, color='salmon')
    ax.set_yticks(np.arange(15, -1, -1))
    ax.set_title(char2name[int2char[i]])
plt.tight_layout()
plt.show()
```

```python
sideviews_mean = {k:np.sum(np.arange(15, -1, -1) * (v / np.sum(v))) for k, v in sideviews.items()}
sideviews_mean
```

```python
df = pd.DataFrame(np.array([list(sideviews_mean.values())]))
df.columns = list(char2name.values())
df.head()

categories = list(df)
values = df.mean().values.flatten().tolist()
values += values[:1] # repeat the first value to close the circular graph
angles = [n / float(len(categories)) * 2 * pi for n in range(len(categories))]
angles += angles[:1]

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(5, 5),
                       subplot_kw=dict(polar=True))

plt.xticks(angles[:-1], categories, color='black')
plt.yticks(color='grey')
ax.set_rlabel_position(30)
ax.set_title('Mean heights of different tiles for training data', y=1.08)
 
ax.plot(angles, values, linewidth=1, linestyle='solid')
ax.fill(angles, values, 'skyblue', alpha=0.5)

plt.show()
```

## Heatmap approach

```python
heatmaps = {i:[] for i in np.unique(smb_int)}
unique_indexes = np.unique(smb_int)
for chunk in tqdm_notebook(smb_int):
    for i in unique_indexes:
        chunk_mod = chunk == i
        heatmaps[i].append(chunk_mod)
heatmaps = {k:np.array(v).mean(axis=0) for k, v in heatmaps.items()}
```
